```python
name = "Zameer Ali Tunio"
father_name = "Shahmir Ali"
course = "AI batch-4"
institute = "Saylani Mass IT Traing Center Karachi"

card = f"""
Name : {name}
Father Name : {father_name}
course: {course}
Institute : {institute}
"""

# 1. capitalize()
print(card.capitalize())

# 2. casefold()
print(card.casefold())

# 3. center()
print(card.center(50, '-'))

# 4. count()
print(card.count('a'))

# 5. encode()
print(card.encode())

# 6. endswith()
print(card.endswith('Karachi'))

# 7. expandtabs()
print(card.expandtabs(tabsize=4))

# 8. find()
print(card.find('batch'))

# 9. format()
print('My name is {} and I am enrolled in {}'.format(name, course))

# 10. format_map()
info = {'name': name, 'course': course}
print('My name is {name} and I am enrolled in {course}'.format_map(info))

# 11. index()
print(card.index('batch'))

# 12. isalnum()
print(card.isalnum())

# 13. isalpha()
print(card.isalpha())

# 14. isascii()
print(card.isascii())

# 15. isdecimal()
print(card.isdecimal())

# 16. isdigit()
print(card.isdigit())

# 17. isidentifier()
print(card.isidentifier())

# 18. islower()
print(card.islower())

# 19. isnumeric()
print(card.isnumeric())

# 20. isprintable()
print(card.isprintable())

# 21. isspace()
print(card.isspace())

# 22. istitle()
print(card.istitle())

# 23. isupper()
print(card.isupper())

# 24. join()
name_list = name.split()
print('-'.join(name_list))

# 25. ljust()
print(card.ljust(50, '*'))

# 26. lower()
print(card.lower())

# 27. lstrip()
print(card.lstrip())

# 28. maketrans()
table = card.maketrans('a', '1')
print(card.translate(table))

# 29. partition()
print(card.partition('course'))

# 30. replace()
print(card.replace('Karachi', 'Pakistan'))

# 31. rfind()
print(card.rfind('batch'))

# 32. rindex()
print(card.rindex('batch'))

# 33. rjust()
print(card.rjust(50, '*'))

# 34. rpartition()
print(card.rpartition('course'))

# 35. rstrip()
print(card.rstrip())

# 36. split()
print(card.split())

# 37. rsplit()
print(card.rsplit())

# 38. splitlines()
print(card.splitlines())

# 39. startswith()
print(card.startswith('Name'))

# 40. strip()
print(card.strip())

# 41. swapcase()
print(card.swapcase())

# 42. title()
print(card.title())

# 43. translate()
table = card.maketrans('AEIOU', '12345')
print(card.translate(table))

# 44. upper()
print(card.upper())

# 45. zfill()
print(card.zfill(100))
```
